```python
X = [[1,2],[2,3],[3,3],[2,1],[3,2]]
y = [1,1,1,-1,-1]
learning_rate = 0.001
lambda_param = 0.01
n_iters = 1000
```

```python
w = [0.0, 0.0]
b = 0.0
iteration = 0
while iteration < n_iters:
    i = 0
    while i < len(X):
        x_i = X[i]
        if y[i] > 0:
            label = 1
        else:
            label = -1
        dot = x_i[0]*w[0] + x_i[1]*w[1] - b
        condition = (label * dot >= 1)
        i = i + 1
    iteration = iteration + 1
    if condition:
      w[0] = w[0] - learning_rate * (2 * lambda_param * w[0])
      w[1] = w[1] - learning_rate * (2 * lambda_param * w[1])
    else:
      w[0] = w[0] - learning_rate * (2 * lambda_param * w[0] - x_i[0]*label)
      w[1] = w[1] - learning_rate * (2 * lambda_param * w[1] - x_i[1]*label)
      b = b - learning_rate * label
```

```python
predictions = []
for i in range(len(X)):
    dot = X[i][0]*w[0] + X[i][1]*w[1] - b
    predictions.append(1 if dot >= 0 else -1)
print("Weights:", w)
print("Bias:", b)
print("Predictions:", predictions)
```
